```python
#INPUT TAKING
n=open('/content/drive/MyDrive/A. Shihab work station/CSE422/ass1/Input_file.txt','r')
dict={}
for i in range(20):
  hold=n.readline().split()
  #dict[node]=[hurestic, [neibours]] || 0- node name, 1- hurestics
  dict[hold[0]]=[int(hold[1]),[]]
  for j in range(2,len(hold),2):
    #appends on the node's neibours list  || #{Node: [huristic,[(neibour, cost from parent)]]}
    dict[hold[0]][1].append((hold[j],int(hold[j+1])))
# sample print(dict["Arad"])==[366, [('Zerind', 75), ('Timisoara', 118), ('Sibiu', 140)]]

start='Craiova'
end='Timisoara'

#priority queue
from queue import PriorityQueue
q = PriorityQueue()
#initial queue
temp= dict[start]
#take h(n), name, g(n) |and| enqueue = [total cost, [node name],cost]
q.put([temp[0],[start],0])

#A function to print the final result
def pathsearch(array,cost, start, end):
  store=""
  for i in range(len(array)):
    if i!=len(array)-1:
      store+=array[i]+", "
    else:
      store+=array[i]
  return f'Optimal Path from {start} to {end} is {store} with cost of {cost} km'

visited=[]
path_found=False
while not q.empty():
  deque= q.get()
  #deque = [total cost, [node name],cost]
  node=deque[1][len(deque[1])-1]
  cost=deque[2]
  visited.append(node)
  #checker
  if node==end:
    path_found=True
    #using self made function to print final result
    print(pathsearch(deque[1],cost, start, end))
    break
  #dict[node]= [h(n),[(neibour, cost)]]
  for tup in dict[node][1]:

    if tup[0] not in visited:
      #cost= ac.cost of dequed pat
      pathcost=tup[1]+cost
      huristic=dict[tup[0]][0]
      totalcost=huristic+ pathcost
      #making copy of path upto parent node
      copy=deque[1].copy()
      new=[totalcost,copy,pathcost]
      #addtion of newest path
      new[1].append(tup[0])
      q.put(new)
if path_found==False:
  print("NO PATH FOUND")
```
