```python
#WORKING
import random
import copy

#Taking the input
input=['5','l 100','l 450','d 500','l 7923','d 9055']
#input= ['7', 'l 120','l 289','d 475','l 195','d 6482','l 160','d 935']

#len= total tranzaction possible
leng=int(input[0])

#array= all tranzactions input
array=[]
for i in range(1,leng+1):
  if 'l' in input[i]:
    temp= input[i].split()
    array.append(-int(temp[1]))
  elif 'd' in input[i]:
    temp= input[i].split()
    array.append(int(temp[1]))
print("INP",array)



#FUNCTION: to check if all 0 in an array
def ALL_0_checker(arr):
  f=True
  for i in arr:
    if i==1:
      f=False
  return f

#populating the states in 0/1 randomizing (0 stating dont pick 1 as pick)
popu=[]

#generating 10 possibilities
i=0

while i!=10:
  #t_arr= [0,1 combinations array--populations]
  t_arr=[]

  #each having 7 tranzactions to pick from
  for idx in range(leng):
    binary = random.randint(0, 1)
    t_arr.append(binary)

  f=ALL_0_checker(t_arr)
  if f==True:
    #print("POPU",popu)
    #print(t_arr)
    pass
  elif t_arr in popu:
    pass
  else:
    popu.append(t_arr)
    i+=1
#print("POPU2",popu)

#FUNCTION: LOW TO HIGH PROBABILITY TAKER
def Low_to_high_prob_maker(a):
  b=[]
  t=0
  for i in a:
    t+=i
  for j in a:
    b.append(j/t)
  return (b)


#FUNCTION: FITNESS probability calculation
def FITNESS(popu,array):
  #Fitness solid value of each population
  fitness=[]

  #Taking each population combination in b_list
  for b_list in popu:
    #0*value=-(don't pick), 1*value=pick
    t_sum=0

    for i in range(leng):
      x=b_list[i]
      y=array[i]
      t_sum+=(x*y)
    fitness.append(abs(t_sum))
  #print(fitness)

  if 0 in fitness:
    print(fitness)
    print("POPUUUUUUUUU",popu)
    return "GOAL FOUND"

  #Total sum of all fitness for probability calc
  total = 0
  for ele in fitness:
      total += ele
  #print(total)

  fit_prob=[]
  for i in fitness:
    low=(i)/total
    fit_prob.append(1/low)

  return Low_to_high_prob_maker(fit_prob)



#FUNCTION: Cumalitive probability calculator
def Cumalitive(list):
  new=[list[0]]
  for i in range (1,len(list)):
    s=new[i-1]
    s+=list[i]
    new.append(s)
  return new

#FUNCTION: Pick two from crossover randomly using cumalitive prob
def Pick_for_Crossover(cuma):
  r1=random.random()
  r2=random.random()

  idx_1= None
  for i in range(len(cuma)):
    if cuma[i]>=r1 :
      idx_1=i
      break
  idx_2= None
  for i in range(len(cuma)):
    if cuma[i]>=r2 :
      idx_2=i
      break

  return [idx_1,idx_2]




#Crossover offspring generator for two
def CROSSOVER(p_1,p_2,leng):

  while True:

    idx=random.randint(1, leng-2)
    c1=p_1[:idx]
    c2=p_2[:idx]

    p_1_2=p_1[idx:leng+1]
    p_2_2=p_2[idx:leng+1]

    c1.extend(p_2_2)
    c2.extend(p_1_2)

    check1=ALL_0_checker(c1)
    check2=ALL_0_checker(c2)

    if check1 ==True or  check2 ==True:
      pass
    else:
      return [c1,c2]
      break








#-----------WHILE LOOP STARTS----------------

j=0
result_flag=None
while j<=500:
  #print(j)
  j+=1

  #CROSSOVER

  prob_2=FITNESS(popu,array)

  if prob_2=="GOAL FOUND":
      print(prob_2)
      result_flag=0
      break

  cuma= Cumalitive(prob_2)
  print("cumalitive",cuma)

  picked= Pick_for_Crossover(cuma)
  print("Picked", picked)

  crossed= CROSSOVER(popu[picked[0]],popu[picked[1]],leng)

  check_zero=copy.deepcopy(crossed)

  print("CROSSED",crossed)
  print("CHECK ZERO", check_zero)

  #MUTATION

  for i in range(2):
    setted=0.3
    r_m= random.random()
    if r_m<setted:
      idx_sel= random.randint(0, leng-1)
      new_val=random.randint(0, 1)
      check_zero[i][idx_sel]=new_val
      c_m=ALL_0_checker(check_zero[i])
      print("checker",c_m)
      if c_m==True:
        print("YAYYYYYYY")
        print(crossed)
        pass
      else:
        crossed[i][idx_sel]=new_val
      #mutation check


  popu[picked[0]]=crossed[0]
  popu[picked[1]]=crossed[1]

  print("MUTATION POPU",popu)

if result_flag==0:
  print(0)
else:
  print(-1)



  #print("AFTER CROSSOVER")
```
