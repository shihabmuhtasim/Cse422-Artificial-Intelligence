```python
#WORKING
import random

#Taking the input
#input=['5','l 100','l 450','d 500','l 7923','d 9055']
input= ['7', 'l 120','l 289','d 475','l 195','d 6482','l 160','d 935']

#len= total tranzaction possible
leng=int(input[0])

#array= all tranzactions input
array=[]
for i in range(1,leng+1):
  if 'l' in input[i]:
    temp= input[i].split()
    array.append(-int(temp[1]))
  elif 'd' in input[i]:
    temp= input[i].split()
    array.append(int(temp[1]))
print("INP",array)

#check if combinations add up to zero
def Combination_checker(arr):
  from itertools import combinations

  sample_list = arr
  list_combinations = []

  for r in range(1, len(sample_list) + 1):
    combinations_r = list(combinations(sample_list, r))
    list_combinations.extend(combinations_r)

  #getting all combinations of the values in list then checking if any sum up to 0 or not
  new=[]
  for i in range(len(list_combinations)):
    if i>len(sample_list)-1:
      s=0
      for j in range(len(list_combinations[i])):
        s+=list_combinations[i][j]
      new.append(s)

  if 0 in new:
    flag="TRUE"
  else:
    flag="FALSE"
  return flag
#--------------------


flag=Combination_checker(array)

if flag=="FALSE":
  print(-1)

#--------part 2---------
else:

  #FUNCTION: to check if all 0 in an array
  def ALL_0_checker(arr):
    f=True
    for i in arr:
      if i==1:
        f=False
    return f

  #populating the states in 0/1 randomizing (0 stating dont pick 1 as pick)
  popu=[]

  #generating 10 possibilities
  i=0

  while i!=10:
    #t_arr= [0,1 combinations array--populations]
    t_arr=[]

    #each having 7 tranzactions to pick from
    for idx in range(leng):
      binary = random.randint(0, 1)
      t_arr.append(binary)

    f=ALL_0_checker(t_arr)
    if f==True:
      #print("POPU",popu)
      #print(t_arr)
      pass
    else:
      popu.append(t_arr)
      i+=1
  #print("POPU2",popu)

  #FUNCTION: LOW TO HIGH PROBABILITY TAKER
  def Low_to_high_prob_maker(a):
    b=[]
    t=0
    for i in a:
      t+=i
    for j in a:
      b.append(j/t)
    return (b)


  #FUNCTION: FITNESS probability calculation
  def FITNESS(popu,array):
    #Fitness solid value of each population
    fitness=[]

    #Taking each population combination in b_list
    for b_list in popu:
      #0*value=-(don't pick), 1*value=pick
      t_sum=0

      for i in range(leng):
        x=b_list[i]
        y=array[i]
        t_sum+=(x*y)
      fitness.append(abs(t_sum))
    #print(fitness)

    if 0 in fitness:
      return "GOAL FOUND"

    #Total sum of all fitness for probability calc

    fitness_2=[]
    for i in fitness:
      fitness_2.append(1/i)

    total = 0
    for ele in fitness_2:
        total += ele
    #print(total)

    fit_prob=[]
    for i in fitness_2:
      low=(i)/total
      fit_prob.append(low)

    return fit_prob



  #FUNCTION: Cumalitive probability calculator
  def Cumalitive(list):
    new=[list[0]]
    for i in range (1,len(list)):
      s=new[i-1]
      s+=list[i]
      new.append(s)
    return new

  #FUNCTION: Pick two from crossover randomly using cumalitive prob
  def Pick_for_Crossover(cuma):
    r1=random.random()
    r2=random.random()

    idx_1= None
    for i in range(len(cuma)):
      if cuma[i]>=r1 :
        idx_1=i
        break
    idx_2= None
    for i in range(len(cuma)):
      if cuma[i]>=r2 :
        idx_2=i
        break

    return [idx_1,idx_2]

  #Crossover offspring generator for two
  def CROSSOVER(p_1,p_2,leng):

    idx=random.randint(1, leng-2)
    c1=p_1[:idx]
    c2=p_2[:idx]

    p_1_2=p_1[idx:leng+1]
    p_2_2=p_2[idx:leng+1]

    c1.extend(p_2_2)
    c2.extend(p_1_2)

    return c1,c2



#-----------WHILE LOOP STARTS----------------

  j=0
  while j<=30:
    #print(j)
    j+=1
    #Mutation
    con=0.2
    prob=FITNESS(popu,array)
    if prob=="GOAL FOUND":
      print(prob)
      break
    for i in range(leng):
      if prob[i]>=con:
        idx_sel= random.randint(0, 6)
        new_val=random.randint(0, 1)
        popu[i][idx_sel]=new_val

    #print("NEW POPU AFTER MUTATION:",popu)

    #CROSSOVER

    prob_2=FITNESS(popu,array)

    if prob_2=="GOAL FOUND":
        print(prob_2)
        break

    cuma= Cumalitive(prob_2)
    print("cumalitive",cuma)

    picked= Pick_for_Crossover(cuma)

    crossed= CROSSOVER(popu[picked[0]],popu[picked[1]],leng)

    popu[picked[0]]=crossed[0]
    popu[picked[1]]=crossed[1]

    #print("AFTER CROSSOVER")
```
