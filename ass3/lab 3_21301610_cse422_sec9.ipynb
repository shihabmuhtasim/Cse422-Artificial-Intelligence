```python
import random

id = input("Enter student id of Brac University: ")

#Taking the limit as a list and making it an integer
limit = input("Enter start and end range seperated by space: ").split(" ")
for i in range (len(limit)):
    limit[i]=int(limit[i])

depth = int(id[0]) * 2
hp = int(id[7] + id[6])
b_factor = int(id[2])


class Node:
    #Main constructor for creating tree location and adding child when needed
    def __init__(self, val):
        self.val = val
        self.child = []
    #adds child to current self node's children list
    def add_child(self, child):
        self.child.append(child)

#Tree creating class
class Tree:
    def __init__(self, depth, b_factor, limit):
        self.root = None
        self.depth = depth
        self.b_factor = b_factor
        self.limit = limit
        self.terminals=[]
    #start tree creation and last recursed root location creator
    def initialize_tree(self):
        self.root = self.Make_tree(self.depth)

    #Main tree creation recursive function
    def Make_tree(self, depth):
        #When reached last node, go back to previous node
        if depth == 0:
            child = Node(random.randint(self.limit[0], self.limit[1]))
            self.terminals.append(child.val)
            return child
        else:
            node = Node(None)
            for i in range(self.b_factor):
                child = self.Make_tree(depth - 1)
                #after getting a node recursed, add that node as a child to this current self node's children and continue loop
                node.add_child(child)

            return node




def Minmax(pos, alpha, beta, maximizing_player, count=0):
    #If a node has no children and no value then it's none
    if pos is None:
        return 0,count
    #when po.val is not none it means it reached the leaf node
    if pos.val is not None:
        return pos.val, count
    # when it's the turn of max player
    if maximizing_player:
        maxEval=float('-inf')
        #po.child is a array so node iterates over it
        for node in pos.child:
          #the next turn is for min so false pass
            eval_val,count=Minmax(node, alpha, beta, False, count)
            #
            maxEval=max(maxEval,eval_val)
            alpha = max(alpha, maxEval)
            count+=1
            if alpha >= beta:
                break
        return maxEval,count
    else:
      #turn for min player
        minEval=float('inf')
        for node in pos.child:
            eval_val,count=Minmax(node, alpha, beta, True, count)
            minEval=min(minEval,eval_val)
            beta = min(beta, minEval)
            count+=1
            if alpha >= beta:
                break
        return minEval,count


my_tree = Tree(depth, b_factor, limit)
my_tree.initialize_tree()

# store the root node for minmax passing
root_node = my_tree.root

attack, counter = Minmax(root_node, float('-inf'), float('inf'), True)

print(f"Depth and Branches ratio is {depth}:{b_factor}")
print(f"Terminal States (leaf node values) are {my_tree.terminals}.")
print(f"Left life(HP) of the defender after maximum damage caused by the attacker is {hp-attack}")
print(f"After Alpha-Beta Pruning Leaf Node Comparisons {counter}")
```
